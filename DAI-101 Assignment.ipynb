```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot: Patient counts by neighborhood
plt.figure(figsize=(10, 8))
neighborhood_counts = data['Neighbourhood'].value_counts().head(10)  # Show top 10 neighborhoods
sns.barplot(x=neighborhood_counts.values, y=neighborhood_counts.index, palette='viridis')
plt.title('Top 10 Neighborhoods with Most Patients')
plt.xlabel('Patient Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
numeric_columns = ['Age', 'Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received']
correlation_matrix = data[numeric_columns].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
(data.groupby('Scholarship').sum()['Handcap']).plot(kind="bar")

# Adding titles and labels
plt.title('average quantiy produced in each product line')
plt.xlabel('Scholarship')
plt.ylabel('Handcap')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Univariate Analysis - Categorical Variables
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
fig.suptitle('Univariate Analysis of Categorical Variables', fontsize=16)

# Gender
sns.countplot(ax=axes[0, 0], x=data['Gender'], palette="Set2")
axes[0, 0].set_title('Gender Distribution')

# Scholarship
sns.countplot(ax=axes[0, 1], x=data['Scholarship'], palette="Set2")
axes[0, 1].set_title('Scholarship Distribution (0 = No, 1 = Yes)')

# Hypertension
sns.countplot(ax=axes[1, 0], x=data['Hipertension'], palette="Set2")
axes[1, 0].set_title('Hypertension Distribution')

# Diabetes
sns.countplot(ax=axes[1, 1], x=data['Diabetes'], palette="Set2")
axes[1, 1].set_title('Diabetes Distribution')

# Alcoholism
sns.countplot(ax=axes[2, 0], x=data['Alcoholism'], palette="Set2")
axes[2, 0].set_title('Alcoholism Distribution')

# No-show
sns.countplot(ax=axes[2, 1], x=data['No-show'], palette="Set2")
axes[2, 1].set_title('No-show Distribution (0 = Attended, 1 = Missed)')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# No-show vs Age
plt.figure(figsize=(10, 6))
sns.boxplot(x='No-show', y='Age', data=data, palette="Set2")
plt.title('No-show vs Age')
plt.show()

# No-show vs SMS_received
plt.figure(figsize=(8, 5))
sns.countplot(x='SMS_received', hue='No-show', data=data, palette="Set2")
plt.title('No-show vs SMS Received')
plt.show()

# No-show vs Hypertension
plt.figure(figsize=(8, 5))
sns.countplot(x='Hipertension', hue='No-show', data=data, palette="Set2")
plt.title('No-show vs Hypertension')
plt.show()

# No-show vs Diabetes
plt.figure(figsize=(8, 5))
sns.countplot(x='Diabetes', hue='No-show', data=data, palette="Set2")
plt.title('No-show vs Diabetes')
plt.show()

# No-show vs Alcoholism
plt.figure(figsize=(8, 5))
sns.countplot(x='Alcoholism', hue='No-show', data=data, palette="Set2")
plt.title('No-show vs Alcoholism')
plt.show()
```
